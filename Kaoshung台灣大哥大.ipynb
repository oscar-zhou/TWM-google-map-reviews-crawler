```python
def catch_review(store_name,reviews_df):
#下滑網頁載入評論
    for i2 in range(10):
        pane = browser.find_element(By.XPATH, '//*[@id="pane"]/div/div[1]/div/div/div[2]')
        browser.execute_script("arguments[0].scrollTop = arguments[0].scrollHeight", pane)
        time.sleep(3)
# 獲取網頁原始碼
    soup = Soup(browser.page_source,"lxml")
# 獲取評論資料框架
    all_reviews = soup.find_all(class_ = 'jftiEf L6Bbsd fontBodyMedium')
    x = len(all_reviews)
    for i3 in range(x):
        ar = all_reviews[i3]
        name = ar.find(class_ = "d4r55").text
        star = ar.find(class_ = "kvMYJc").get('aria-label')[1]
        star = int(star)
        content = ar.find(class_ = "wiI7pd").text
        review_time = ar.find(class_ = "rsqaWe").text
        try:
            local = ar.find(class_ = "RfnDt").find_all('span')[0]
            local = str(local)
            style = "display:none"
            if (local.find(style)==-1):
                local_flag = 1
            else:
                local_flag = 0
        except:
            loacl_flag = 0
        try:
            review_num = ar.find(class_ = "RfnDt").find_all('span')[1].text
            review_num = review_num.replace('則評論','')
            review_num = review_num.replace(' ','')
            review_num = int(review_num)
        except:
            review_num = 0
        
#評論長度 空值為3
        if len(content)>3 :
            one_review = {
                "name":name,
                "local":local_flag,
                "review_num":review_num,
                "star":star,
                "review":content,
                "review_time":time_catch(review_time),
                "store_name":store_name}
            reviews_df = reviews_df.append(one_review,ignore_index=True)
    return reviews_df
```
